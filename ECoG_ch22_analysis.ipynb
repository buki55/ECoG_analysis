```python
from numpy import *
from scipy import *
from scipy.io import loadmat
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = loadmat('ECoG_ch22.mat')
```

```python
dataset
```

```python
ecog_data = dataset['ECoGData_ch22']
ecog_signal = ecog_data[0] / 1000
ecog_time = arange(0, ecog_data[0].shape[0])
ecog_time = ecog_time / 1000
```

```python
plt.plot(ecog_time, ecog_signal, linewidth=0.5)
plt.xlabel('Time')
plt.ylabel('Signal')
plt.grid(True)
```

```python
plt.plot(ecog_time, ecog_signal)
plt.xlabel('Time')
plt.ylabel('Signal')
plt.xlim(0, 1)
plt.grid(True)
```

```python
from scipy import signal
# サンプリング周波数
fs = 1000.0
f,t,Sxx = signal.spectrogram(ecog_signal, fs, nperseg=256)
```

```python
plt.pcolormesh(t, f, Sxx, vmax=1e-5)
plt.xlabel('Time[sec]')
plt.ylabel('frequences[Hz]')
```

```python
plt.pcolormesh(t, f, Sxx, vmax=1e-4)
plt.xlabel('Time[sec]')
plt.ylabel('frequences[Hz]')
plt.xlim(0,100)
```
